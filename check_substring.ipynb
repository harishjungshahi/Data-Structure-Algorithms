```python
def is_substring(s,t):
  n=len(s)
  m=len(t)

  if m>n:
    return False

  i = 0
  while i<=n-m:
    j=0
    while j<m and s[i+j]==t[j]:
      j+=1

    if j==m:
      return True

    i += 1
  return False


print(is_substring("hello world", "world"))
print(is_substring("hello world", "worlds"))
```
