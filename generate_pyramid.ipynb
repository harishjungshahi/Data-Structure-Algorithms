```python
def generate_pyramid(n):
    pyramid = []

    for i in range(n):
        spaces = ' ' * (n - i - 1)
        stars = '*' * (2 * i + 1)
        pyramid.append(spaces + stars + spaces)

    return pyramid
for row in generate_pyramid(5):
    print(row)
```
