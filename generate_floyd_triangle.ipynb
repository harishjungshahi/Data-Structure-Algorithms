```python
def generate_floyd_triangle(n):
    triangle = []
    num = 1   # starting number

    for i in range(1, n + 1):
        row = []

        for j in range(i):
            row.append(str(num))
            num += 1   # increase number

        triangle.append(' '.join(row))

    return triangle

for row in generate_floyd_triangle(5):
    print(row)
```
