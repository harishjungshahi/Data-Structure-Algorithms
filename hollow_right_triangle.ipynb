```python
def generate_hollow_right_triangle(n):
    """
    Function to generate a hollow right-angled triangle pattern as a list of strings.

    Parameters:
    n (int): Number of rows in the triangle

    Returns:
    list: List of strings representing the hollow right-angled triangle
    """
    triangle = []

    for i in range(1, n + 1):
        if i == 1:
            # First row has 1 star
            triangle.append('*')
        elif i == 2:
            # Second row has 2 stars
            triangle.append('**')
        elif i == n:
            # Last row has n stars
            triangle.append('*' * i)
        else:
            # Middle rows: star + spaces + star
            triangle.append('*' + ' ' * (i - 2) + '*')

    return triangle

for row in generate_hollow_right_triangle(5):
    print(row)
```
