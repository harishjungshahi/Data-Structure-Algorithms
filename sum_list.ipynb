```python
def sum_list(numbers):
    # Initialize a variable to store the sum
    total = 0

    # Iterate through each number in the list
    for num in numbers:
        # Add the current number to the total
        total += num

    # Return the final sum
    return total
print(sum_list([1, 2, 3, 4, 5]))
```
