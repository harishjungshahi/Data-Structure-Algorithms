```python
def generate_inverted_triangle(n):
    triangle = []

    for i in range(n, 0, -1):  # from n down to 1
        triangle.append('*' * i)

    return triangle
print(generate_inverted_triangle(5))


for row in generate_inverted_triangle(5):
    print(row)
```
