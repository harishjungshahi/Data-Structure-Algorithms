```python
def are_equal_strings(s,t):
  len_s = 0
  len_t = 0

  len_s = len(s)
  len_t = len(t)

  if len(s)!=len(t):
    return False

  for i in range(len_s):
    if s[i]!=t[i]:
      return False

  return True

print(are_equal_strings('Harish', 'Harish'))
print(are_equal_strings('Harish', 'Ashish'))
```

```python
def reverse_string(s):
  reversed_str = ' '

  for i in range(len(s)-1,-1,-1):
    reversed_str += s[i]

  return reversed_str

print(reverse_string('Hello'))
```

```python
def count_vowels(s):
  vowels = 'aeiouAEIOU'
  count = 0

  for char in s:
    if char in vowels:
      count +=1
  return count

print(count_vowels('Harish Jung Shahi'))
```

```python
def print_vowels(s):
  vowels='aeiouAEIOU'
  print = []

  for char in s:
    if char in vowels:
      print.append(char)

  return print
print(print_vowels('Harish Jung Shahi'))
```

```python
def is_palindrome(s):
  normalized_str = ''.join(char.lower() for char in s if char.isalnum())

  return normalized_str == normalized_str[::-1]

print(is_palindrome('madam'))
```

```python
def count_words(s):
  count = 0
  in_word = False

  for char in s:
    if char != ' ':
      if not in_word:
        count += 1
        in_word = True

    else:
      in_word = False

  return count
print(count_words('Harish is awesome'))
```
