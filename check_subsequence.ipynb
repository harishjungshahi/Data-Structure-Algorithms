```python
def is_subsequence(s,t):
  i=0
  j=0

  while i<len(s) and j<len(t):
    if s[i]==t[j]:
      j +=1
    i+=1

  if j==len(t):
    return True
  else:
    return False

print(is_subsequence("abcde","abe"))
print(is_subsequence("abcde","aec"))
```
