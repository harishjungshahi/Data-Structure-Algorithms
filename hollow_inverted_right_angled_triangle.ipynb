```python
def generate_hollow_inverted_right_angled_triangle(n):
    triangle = []

    for i in range(n, 0, -1):  # from n down to 1
        if i == n:
            # First row (full stars)
            triangle.append('*' * n)
        elif i == 1:
            # Last row (single star)
            triangle.append('*')
        else:
            # Middle rows (hollow)
            triangle.append('*' + ' ' * (i - 2) + '*')

    return triangle

for row in generate_hollow_inverted_right_angled_triangle(5):
    print(row)
```
