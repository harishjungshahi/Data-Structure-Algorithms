```python
def remove_duplicates(s):
  result = " "

  for char in s:
    found = False

    # Check if character already exists in result

    i = 0
    while i < len(result):
      if result[i]==char:
        found = True
        break
      i +=1

    # If not found, add to result
    if not found:
      result +=char

  return result


print(remove_duplicates("programming"))
print(remove_duplicates("Hello WOrld!"))
```
