```python
def is_anagram(s,t):
  if len(s)!=len(t):
    return False
  count = [0]*26

  i = 0
  while i < len(s):
    index = ord(s[i]) - ord('a')
    count[index] +=1
    i+=1

  i = 0
  while i < len(t):
    index =ord(t[i])-ord('a')
    count[index] -=1
    i +=1

  i = 0
  while i < 26:
    if count[i]!=0:
      return False
    i+=1
  return True

print(is_anagram('anagram','nagaram'))
print(is_anagram('rat','cat'))

```
