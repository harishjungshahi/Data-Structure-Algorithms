```python
def are_equal_strings(s,t):
  len_s = 0
  len_t = 0

  while len_s < len(s):
    len_s +=1
  while len_t < len(t):
    len_t +=1

  if len_s != len_t:
    return False

  for i in range(len_s):
    if s[i] != t[i]:
      return False

  return True

print(are_equal_strings("hello", "hello"))
```
