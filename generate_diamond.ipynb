```python
def generate_right_aligned_triangle(n):
    """
    Function to return a right-angled triangle pattern of '*' aligned to the right.

    Parameters:
    n (int): Number of rows of the triangle.

    Returns:
    list: A list of strings representing the right-aligned triangle.
    """
    triangle = []

    for i in range(1, n + 1):
        spaces = ' ' * (n - i)  # spaces to push stars to the right
        stars = '*' * i          # stars for the current row
        triangle.append(spaces + stars)

    return triangle
for row in generate_right_aligned_triangle(4):
    print(row)
```
