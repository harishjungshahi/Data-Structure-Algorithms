```python
def generate_diamond(n):
    """
    Function to return a diamond pattern as a list of strings.

    Parameters:
    n (int): Number of rows in the upper half of the diamond.

    Returns:
    list: A list of strings representing the diamond pattern.
    """
    diamond = []

    # Upper part (pyramid)
    for i in range(n):
        spaces = ' ' * (n - i - 1)
        stars = '*' * (2 * i + 1)
        diamond.append(spaces + stars + spaces)

    # Lower part (inverted pyramid)
    for i in range(n - 2, -1, -1):
        spaces = ' ' * (n - i - 1)
        stars = '*' * (2 * i + 1)
        diamond.append(spaces + stars + spaces)

    return diamond

for row in generate_diamond(3):
    print(row)
```
