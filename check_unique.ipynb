```python
def check_unique(lst):
    seen = []

    for num in lst:
        if num in seen:
            return False
        seen.append(num)

    return True

print(check_unique([1, 2, 3, 4, 5]))
print(check_unique([1, 2, 3, 3, 4, 5]))
```
