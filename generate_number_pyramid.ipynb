```python
def generate_number_pyramid(n):
    pyramid = []

    for i in range(1, n + 1):
        spaces = ' ' * (n - i)  # leading spaces
        numbers = ' '.join(str(j) for j in range(1, i + 1))  # numbers from 1 to i
        pyramid.append(spaces + numbers + spaces)

    return pyramid

# 🔹 Call the function
for row in generate_number_pyramid(4):
    print(row)
```
