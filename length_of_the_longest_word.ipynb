```python
def longest_word_length(s):
    max_length = 0
    current_length = 0
    i = 0

    while i < len(s):
        if s[i] != ' ':   # check for space
            current_length += 1
        else:
            if current_length > max_length:
                max_length = current_length
            current_length = 0
        i += 1

    # check last word
    if current_length > max_length:
        max_length = current_length

    return max_length

print(longest_word_length(" The quickest brown fox jumps over the lazy dog"))
```
