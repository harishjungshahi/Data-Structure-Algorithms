```python
def generate_sandglass(n):
    """
    Function to generate a sandglass pattern of '*' as a list of strings.

    Parameters:
    n (int): Number of rows for the top half (widest row = 2n-1 stars)

    Returns:
    list: List of strings representing the sandglass pattern
    """
    sandglass = []

    # Top half (inverted pyramid)
    for i in range(n):
        spaces = ' ' * i
        stars = '*' * (2 * (n - i) - 1)
        sandglass.append(spaces + stars + spaces)

    # Bottom half (normal pyramid), excluding the middle row to avoid duplication
    for i in range(n - 2, -1, -1):
        spaces = ' ' * i
        stars = '*' * (2 * (n - i) - 1)
        sandglass.append(spaces + stars + spaces)

    return sandglass
for row in generate_sandglass(3):
    print(row)
```
