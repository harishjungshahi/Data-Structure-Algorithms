```python
def reverse_string(s):
  reversed_str = ' '

  for i in range(len(s)-1,-1,-1):
    reversed_str += s[i]

  return reversed_str

print(reverse_string('Hello'))
```

```python
def count_vowels(s):
  vowels = 'aeiouAEIOU'
  count = 0

  for char in s:
    if char in vowels:
      count += 1

  return count
print(count_vowels('Harish Jung Shahi'))
```

```python
def print_vowels(s):
  vowels = 'aeiouAEIOU'
  print = [ ]

  for char in s:
    if char in vowels:
      print.append(char)

  return print
print(print_vowels('Harish Jung Shahi'))
```

```python
def count_words(s):
  count = 0
  in_word = False

  for char in s:
    if char != ' ':
      if not in_word:
        count += 1
        in_word = True
    else:
      in_word = False

  return count

print(count_words('Harish is awesome'))



```
